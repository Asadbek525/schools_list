```python
import requests
import json

id = 2
regions = requests.get(f"https://oo.uzedu.uz/List/TumanList?vilID={id}").json()
sz = len(regions)
regions_list = []
for i in range(1, sz):
    regions_list.append({regions[i]['Text'], regions[i]['Value']})
print(regions_list)
```

```python
districts_list = []
for i in range(1, sz):
    region_id = regions[i]['Value']
    districts = requests.get(f"https://oo.uzedu.uz/List/TumanList?vilID={region_id}").json()
    sz1 = len(districts)
    for j in range(1, sz1):
        districts_list.append({"viloyat":regions[i]['Text'],"tuman":districts[j]['Text'], "tuman_id": districts[j]['Value']})

districts_list = districts_list[:-1]

for i in range(0, len(districts_list)):
    print(districts_list[i])
```

```python
import csv
schools_list = []

for district in districts_list:
    district_id = district['tuman_id']
    schools = requests.get(f"https://oo.uzedu.uz/List/MaktabList?TumanID={district_id}").json()
    sz2 = len(schools)
    for i in range(1, sz2):
        schools_list.append({"viloyat":district['viloyat'], "tuman":district['tuman'], "maktab":schools[i]['Text']})

with open('maktablar.json', 'w',encoding='UTF8', newline='') as file:
    json.dump(schools_list, file, ensure_ascii=False)
```
